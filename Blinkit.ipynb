```python
import pandas as pd
import numpy as np

orders = pd.read_csv('blinkit_data/blinkit_orders.csv')
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
orders['promised_delivery_time'] = pd.to_datetime(orders['promised_delivery_time'])
orders['actual_delivery_time']   = pd.to_datetime(orders['actual_delivery_time'])
```

```python
orders.info()
```

```python
orders['order_month'] = orders['order_date'].dt.month
orders['order_day']   = orders['order_date'].dt.day_name()
orders['order_hour']  = orders['order_date'].dt.hour
```

```python
orders[['order_date', 'order_month', 'order_day', 'order_hour']].head()
```

```python
orders.to_csv('blinkit_data/cleaned_orders.csv', index=False)
print("Orders saved!")
```

```python
customers = pd.read_csv('blinkit_data/blinkit_customers.csv')
```

```python
customers['registration_date'] = pd.to_datetime(customers['registration_date'])
```

```python
customers['phone'] = customers['phone'].astype(str)
```

```python
customers['tenure_days'] = (pd.Timestamp('today') - customers['registration_date']).dt.days
```

```python
customers[['customer_name', 'registration_date', 'phone', 'tenure_days']].head()
```

```python
customers.to_csv('blinkit_data/cleaned_customers.csv', index=False)
print("Customers saved!")
```

```python
delivery = pd.read_csv('blinkit_data/blinkit_delivery_performance.csv')
delivery.isnull().sum()
```

```python
delivery['promised_time'] = pd.to_datetime(delivery['promised_time'])
delivery['actual_time'] = pd.to_datetime(delivery['actual_time'])
delivery['reasons_if_delayed'] = delivery['reasons_if_delayed'].fillna('No Issue')
```

```python
delivery['promised_time'] = pd.to_datetime(delivery['promised_time'])
delivery['actual_time'] = pd.to_datetime(delivery['actual_time'])
```

```python
delivery.isnull().sum()
```

```python
delivery.to_csv('blinkit_data/cleaned_delivery.csv', index=False)
print("Delivery saved!")
```

```python
feedback = pd.read_csv('blinkit_data/blinkit_customer_feedback.csv')
feedback.isnull().sum()
```

```python
feedback['feedback_text'] = feedback['feedback_text'].fillna('No Comment')
```

```python
def rating_label(r):
    if r == 5:   return 'Excellent'
    elif r == 4: return 'Good'
    elif r == 3: return 'Neutral'
    elif r == 2: return 'Poor'
    else:        return 'Very Poor'

feedback['rating_label'] = feedback['rating'].apply(rating_label)
```

```python
print(feedback.isnull().sum())
print(feedback['rating_label'].value_counts())

feedback.to_csv('blinkit_data/cleaned_feedback.csv', index=False)
print("Feedback saved!")
```

```python
marketing = pd.read_csv('blinkit_data/blinkit_marketing_performance.csv')
marketing.isnull().sum()
marketing.head()
```

```python
marketing['date'] = pd.to_datetime(marketing['date'])
```

```python
marketing['ctr'] = (marketing['clicks'] / marketing['impressions'] * 100).round(2)

marketing['cvr'] = (marketing['conversions'] / marketing['clicks'] * 100).round(2)

marketing['cpc'] = (marketing['spend'] / marketing['clicks']).round(2)

marketing['cpa'] = (marketing['spend'] / marketing['conversions']).round(2)
```

```python
marketing[['channel', 'ctr', 'cvr', 'cpc', 'cpa']].head()
```

```python
marketing.to_csv('blinkit_data/cleaned_marketing.csv', index=False)
print("Marketing saved!")
```

```python
inventory = pd.read_csv('blinkit_data/blinkit_inventory.csv')
print(inventory.isnull().sum())
print(inventory.head())
print(inventory['date'].head())   # check date format carefully!
```

```python
inventory['date'] = pd.to_datetime(inventory['date'], format='%d-%m-%Y')
```

```python
inventory['damage_rate_pct'] = (
    inventory['damaged_stock'] / inventory['stock_received'] * 100
).round(2)
```

```python
inventory['high_damage_flag'] = inventory['damage_rate_pct'] > 20
print("High damage days:", inventory['high_damage_flag'].sum())
```

```python
print(inventory.isnull().sum())
inventory.tail()
```

```python
inventory.to_csv('blinkit_data/cleaned_inventory.csv', index=False)
print("Inventory saved!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.size'] = 12

print("Ready to visualize!")
```

```python
path = 'blinkit_data/'

orders    = pd.read_csv(path + 'cleaned_orders.csv')
customers = pd.read_csv(path + 'cleaned_customers.csv')
delivery  = pd.read_csv(path + 'cleaned_delivery.csv')
feedback  = pd.read_csv(path + 'cleaned_feedback.csv')
marketing = pd.read_csv(path + 'cleaned_marketing.csv')
inventory = pd.read_csv(path + 'cleaned_inventory.csv')
items     = pd.read_csv(path + 'blinkit_order_items.csv')
products  = pd.read_csv(path + 'blinkit_products.csv')

print("All cleaned files loaded!")
```

  MONTHLY REVENUE TREND

```python
monthly_revenue = orders.groupby('order_month')['order_total'].sum().reset_index()
monthly_revenue.columns = ['Month', 'Revenue']
print(monthly_revenue)
```

```python
plt.figure(figsize=(12, 5))

sns.barplot(data=monthly_revenue, x='Month', y='Revenue', palette='Greens_d')

plt.title('Monthly Revenue Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Revenue (₹)')
plt.tight_layout()
plt.show()
```

Chart 2 — Orders by Day of Week

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

day_counts = orders.groupby('order_day')['order_id'].count().reindex(day_order).reset_index()
day_counts.columns = ['Day', 'Total_Orders']
print(day_counts)
```

```python
plt.figure(figsize=(12, 5))

sns.barplot(data=day_counts, x='Day', y='Total_Orders', palette='Blues_d')

plt.title('Orders by Day of Week', fontsize=16, fontweight='bold')
plt.xlabel('Day')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

Chart 3 — Orders by Hour of Day

```python
hourly = orders.groupby('order_hour')['order_id'].count().reset_index()
hourly.columns = ['Hour', 'Orders']

plt.figure(figsize=(14, 5))
sns.lineplot(data=hourly, x='Hour', y='Orders', marker='o', color='green', linewidth=2.5)

plt.title('Orders by Hour of Day', fontsize=16, fontweight='bold')
plt.xlabel('Hour (0 = Midnight, 12 = Noon)')
plt.ylabel('Number of Orders')
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

 Chart 4 — Revenue by Customer Segment

```python
orders_customers = orders.merge(customers, on='customer_id', how='left')
print(orders_customers[['order_id', 'customer_segment', 'order_total']].head())
```

```python
segment_revenue = orders_customers.groupby('customer_segment')['order_total'].sum().reset_index()
segment_revenue.columns = ['Segment', 'Revenue']
segment_revenue = segment_revenue.sort_values('Revenue', ascending=False)
print(segment_revenue)
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(data=segment_revenue, x='Segment', y='Revenue',
            hue='Segment', palette='Set2', legend=False)

plt.title('Revenue by Customer Segment', fontsize=16, fontweight='bold')
plt.xlabel('Customer Segment')
plt.ylabel('Total Revenue (₹)')
plt.tight_layout()
plt.show()
```

 Chart 5 — Delivery Status Breakdown

```python
delivery_counts = delivery['delivery_status'].value_counts().reset_index()
delivery_counts.columns = ['Status', 'Count']

plt.figure(figsize=(8, 6))
plt.pie(delivery_counts['Count'],
        labels=delivery_counts['Status'],
        autopct='%1.1f%%',
        colors=['#2ecc71', '#e74c3c', '#f39c12'],
        startangle=90,
        explode=[0.05, 0.05, 0.05])

plt.title('Delivery Status Breakdown', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

Chart 6 — Top 10 Delay Reasons

```python
delayed = delivery[delivery['delivery_status'] != 'On Time']

delay_reasons = delayed['reasons_if_delayed'].value_counts().head(10).reset_index()
delay_reasons.columns = ['Reason', 'Count']
print(delay_reasons)
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=delay_reasons, x='Count', y='Reason',
            hue='Reason', palette='Reds_r', legend=False)

plt.title('Top Reasons for Delivery Delays', fontsize=16, fontweight='bold')
plt.xlabel('Number of Delays')
plt.ylabel('Reason')
plt.tight_layout()
plt.show()
```

 Chart 7 — Feedback Rating Distribution

```python
rating_counts = feedback['rating_label'].value_counts().reset_index()
rating_counts.columns = ['Rating', 'Count']

order = ['Excellent', 'Good', 'Neutral', 'Poor', 'Very Poor']
rating_counts = rating_counts.set_index('Rating').reindex(order).reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=rating_counts, x='Rating', y='Count',
            hue='Rating',
            palette=['#27ae60','#2ecc71','#f39c12','#e67e22','#e74c3c'],
            legend=False)

plt.title('Customer Feedback Rating Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
```

```python
sentiment = feedback['sentiment'].value_counts().reset_index()
sentiment.columns = ['Sentiment', 'Count']

plt.figure(figsize=(7, 7))
plt.pie(sentiment['Count'],
        labels=sentiment['Sentiment'],
        autopct='%1.1f%%',
        colors=['#2ecc71', '#e74c3c', '#95a5a6'],
        startangle=90)

plt.title('Customer Sentiment Breakdown', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
channel_perf = marketing.groupby('channel').agg(
    Total_Spend    = ('spend', 'sum'),
    Total_Revenue  = ('revenue_generated', 'sum'),
    Avg_ROAS       = ('roas', 'mean'),
    Avg_CTR        = ('ctr', 'mean'),
    Avg_CVR        = ('cvr', 'mean')
).reset_index().round(2)

print(channel_perf)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=channel_perf, x='channel', y='Avg_ROAS',
            hue='channel', palette='coolwarm', legend=False)

plt.title('Average ROAS by Marketing Channel', fontsize=16, fontweight='bold')
plt.xlabel('Channel')
plt.ylabel('Return on Ad Spend (ROAS)')
plt.axhline(y=1, color='red', linestyle='--', label='Break Even (ROAS=1)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
channel_melted = channel_perf.melt(id_vars='channel',
                                    value_vars=['Total_Spend', 'Total_Revenue'],
                                    var_name='Metric', value_name='Amount')

plt.figure(figsize=(12, 5))
sns.barplot(data=channel_melted, x='channel', y='Amount',
            hue='Metric', palette=['#e74c3c', '#2ecc71'])

plt.title('Spend vs Revenue by Marketing Channel', fontsize=16, fontweight='bold')
plt.xlabel('Channel')
plt.ylabel('Amount (₹)')
plt.tight_layout()
plt.show()
```
